# Parte 2: Selección y limpieza de los Datos en Python

```python
import pandas as pd
import matplotlib as plt
path = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'
df = pd.read_csv(path)
df
```

```python
total_rows_df = df.ID.count()
total_rows_df
```

#### Renombro las columnas para entender hacer match con la descripcion y trabajar mejor con los datos

```python

df.columns = ['ID', 'given_credit', 'gender', 'education', 'marital_status', 'age', 'repayment_status_sept', 'repayment_status_aug', 
       'repayment_status_jul', 'repayment_status_jun', 'repayment_status_may', 'repayment_status_apr',
       'bill_statemen_sep', 'bill_statemen_aug', 'bill_statemen_jul', 
       'bill_statemen_jun', 'bill_statemen_may', 'bill_statemen_apr',
       'paid_sep', 'paid_aug', 'paid_jul', 'paid_jun', 'paid_may',
       'paid_apr', 'creditable']
df
```

### Reviso si existen datos nulos/ vacíos

```python
df.isnull().values.any()
```

```python
df.isna().any()
```

```python
df.isnull().sum(axis = 0)
```

### Se observa que tenemos datos con genero/educacion/marital nulo como información de usuario empezaremos analizando el que tiene genero nulo

```python
gender_null = df[df['gender'].isna()]
gender_null.index
education_null = df[df['education'].isna()]
education_null
marital_status_null = df[df['marital_status'].isna()]
marital_status_null
```

### Se observa que en los 3 casos tienen la mayoria datos en **NaN** entonces podemos eliminarlos ya que no nos causan ruido

```python
df_copy = df.drop(df.index[[gender_null.index[0],education_null.index[0]]])
df_copy.isnull().sum(axis = 0)
```

### De acuerdo a la descripción de los datos la clase creditable es el resultado de la red neuronal que definio si el usuario es confiable para un credito o no, podemos eliminar los 3 datos que son nulos a este resultado

```python
credible = df_copy[df_copy['creditable'].isna()]
credible
```

```python
df_copy = df_copy.drop(credible.index)
df_copy.isnull().sum(axis = 0)
```

### Para los datos de edad nos podemos basar en la media de la edad de acuerdo al credito obtenido

```python
age = df_copy[df_copy['age'].isna()]
age.given_credit
```

```python
media_edad = df_copy.loc[(df_copy.given_credit < 500000) & (df_copy.given_credit > 140000)].age.mean()
```

### Remplazamos con la media de edad

```python
df_copy.age.fillna(value = int(media_edad), 
                    inplace = True)
df_copy.isnull().sum(axis = 0)
```

### Una vez hecho esta limpieza procedemos a revisar los nulos con mas datos **repayment_status_may** aquí procedo a poner la moda entre los meses de septiembre a abril simulando un comportamiento de "atrasos de pago"

```python
nulos_faltantes = df_copy[df_copy['repayment_status_may'].isna()]
nulos_faltantes
```

### Se observa que tambien tienen bastantes campos en Nulo procedemos a borrarlos para que no nos causen ruido

```python
df_copy.dropna(inplace = True)
df_copy
```

```python
total_rows_df_final = df_copy.ID.count()
total_rows_df_final
```

```python
df_copy.describe()
```

```python
data_rescatada = (total_rows_df_final * 100) / total_rows_df
print(f'{data_rescatada}%')
```
